```python
import plotly.graph_objects as go

def plot_candlestick_chart(filename):
    # Load data from CSV
    df = pd.read_csv(filename)
    df['Date'] = pd.to_datetime(df['Date'])  # Ensure Date is in datetime format

    # Create Candlestick Chart
    fig = go.Figure()

    # Add Candlestick trace
    fig.add_trace(go.Candlestick(
        x=df['Date'],
        open=df['O'],
        high=df['H'],
        low=df['L'],
        close=df['C'],
        name='OHLC',
        increasing_line_color='green',
        decreasing_line_color='red'
    ))

    # Add 50-MA trace
    fig.add_trace(go.Scatter(
        x=df['Date'],
        y=df['50_MA'],
        mode='lines',
        name='50 MA',
        line=dict(color='blue', width=2)
    ))

    # Layout settings
    fig.update_layout(
        title=f'Candlestick Chart for {filename}',
        xaxis_title='Date',
        yaxis_title='Price',
        xaxis_rangeslider_visible=False,
        template="plotly_dark"
    )

    # Show the plot
    fig.show()

# Call the function
plot_candlestick_chart(filename)
```

```python
import MetaTrader5 as mt5
import pandas as pd
import time

def initialize_mt5():
    if not mt5.initialize():
        print("MT5 initialization failed")
        return False
    print("MT5 initialized successfully")
    return True

def get_ohlc_data(symbol, timeframe, bars=45):
    """Fetch OHLC data for a given symbol and timeframe."""
    print(f"Fetching {bars} candles for {symbol} on timeframe {timeframe}...")
    rates = mt5.copy_rates_from_pos(symbol, timeframe, 0, bars)
    if rates is None:
        print(f" Failed to retrieve data for {symbol}")
        return None
    df = pd.DataFrame(rates)
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df.rename(columns={'open': 'O', 'high': 'H', 'low': 'L', 'close': 'C'}, inplace=True)
    df['9_MA'] = df['C'].rolling(window=9).mean()
    print(f" Data fetched successfully for {symbol}")
    return df

def get_trend(symbol):
    """Determine trend using the 9-MA on the 1-hour timeframe."""
    df = get_ohlc_data(symbol, mt5.TIMEFRAME_H1, bars=10)  # Use last 10 candles
    if df is not None and len(df) > 9:
        trend = "Bullish" if df['C'].iloc[-1] > df['9_MA'].iloc[-1] else "Bearish"
        print(f" Trend for {symbol}: {trend} (Close: {df['C'].iloc[-1]}, 9-MA: {df['9_MA'].iloc[-1]})")
        return trend
    print(f" Not enough data to determine trend for {symbol}")
    return None

def detect_engulfing(df, trend):
    """Detect bullish or bearish engulfing pattern."""
    if len(df) < 2:
        print(" Not enough candles to check for engulfing pattern")
        return None, None, None

    prev_open, prev_close = df.iloc[-2]['O'], df.iloc[-2]['C']
    curr_open, curr_close = df.iloc[-1]['O'], df.iloc[-1]['C']

    if trend == "Bullish" and curr_close > curr_open and curr_close > prev_open and curr_open < prev_close:
        print(" Bullish Engulfing detected!")
        return "BUY", curr_close, min(df.iloc[-2]['L'], df.iloc[-1]['L'])
    elif trend == "Bearish" and curr_close < curr_open and curr_close < prev_open and curr_open > prev_close:
        print(" Bearish Engulfing detected!")
        return "SELL", curr_close, max(df.iloc[-2]['H'], df.iloc[-1]['H'])
    
    print(" No engulfing pattern found")
    return None, None, None

def place_order(symbol, order_type, entry_price, sl, tp):
    """Place a market order with calculated SL and TP."""
    tick = mt5.symbol_info_tick(symbol)
    price = tick.ask if order_type == "BUY" else tick.bid  # Use current market price

    print(f" Placing {order_type} order on {symbol} at {price} | SL: {sl} | TP: {tp}")

    request = {
        "action": mt5.TRADE_ACTION_DEAL,
        "symbol": symbol,
        "volume": 0.1,
        "type": mt5.ORDER_TYPE_BUY if order_type == "BUY" else mt5.ORDER_TYPE_SELL,
        "price": price,
        "sl": sl,
        "tp": tp,
        "deviation": 10,
        "magic": 123456,
        "comment": "AutoTrade",
        "type_time": mt5.ORDER_TIME_GTC,
        "type_filling": mt5.ORDER_FILLING_IOC,
    }
    
    result = mt5.order_send(request)
    if result.retcode == mt5.TRADE_RETCODE_DONE:
        print(f" {order_type} order placed successfully on {symbol} at {price}")
    else:
        print(f" Trade failed for {symbol}: {result.comment}")

def check_positions():
    """Check open positions and print their profit/loss."""
    positions = mt5.positions_get()
    if positions:
        print("\n Checking open positions:")
        for pos in positions:
            symbol = pos.symbol
            order_type = "BUY" if pos.type == 0 else "SELL"
            profit = pos.profit
            print(f" {symbol} | {order_type} | Entry: {pos.price_open} | SL: {pos.sl} | TP: {pos.tp} | P/L: {profit:.2f}")
    else:
        print("\n No open positions.")

def main():
    if not initialize_mt5():
        return
    
    symbols = ["EURUSD", "GBPUSD", "USDJPY", "XAUUSD", "USDCHF"]
    
    while True:
        for symbol in symbols:
            print(f"\n Analyzing {symbol}...")
            
            # Check trend
            trend = get_trend(symbol)
            if trend is None:
                print(f" Skipping {symbol}, unable to determine trend.")
                continue

            # Fetch 15-min timeframe data
            df = get_ohlc_data(symbol, mt5.TIMEFRAME_M15, bars=45)
            if df is None:
                print(f" Skipping {symbol}, no data available.")
                continue

            # Look for engulfing patterns
            order_type, entry_price, sl = detect_engulfing(df, trend)
            if order_type:
                tp = entry_price + (entry_price - sl) * 2 if order_type == "BUY" else entry_price - (sl - entry_price) * 2
                place_order(symbol, order_type, entry_price, sl, tp)
            else:
                print(f" No valid trade setup found for {symbol}")

        check_positions()  # Print running P/L every loop
        print("\n Waiting for the next cycle...\n")
        time.sleep(60)  # Check every minute

if __name__ == "__main__":
    main()
```

```python
import MetaTrader5 as mt5
import pandas as pd
import numpy as np
import time
import datetime

# Connect to MetaTrader 5
if not mt5.initialize():
    print("MT5 initialization failed!")
    quit()

# Function to fetch OHLC data with VWAP and Order Blocks
def get_ohlc_data(symbol, timeframe, bars=1000):
    rates = mt5.copy_rates_from_pos(symbol, timeframe, 0, bars)
    if rates is None or len(rates) == 0:
        print(f"Failed to retrieve data for {symbol}")
        return None
    df = pd.DataFrame(rates)
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df.rename(columns={'time': 'Date', 'open': 'O', 'high': 'H', 'low': 'L', 'close': 'C'}, inplace=True)
    df['VWAP'] = (df['O'] + df['H'] + df['L'] + df['C']) / 4  # Simple VWAP approximation
    df['50_MA'] = df['C'].rolling(window=50).mean()
    return df

# Function to detect Institutional Order Blocks
def detect_order_blocks(df):
    ob_list = []
    if df is None or len(df) < 3:
        return ob_list
    for i in range(1, len(df) - 1):
        if df.loc[i - 1, 'H'] < df.loc[i + 1, 'L']:  # Bullish OB
            ob_list.append((df.loc[i, 'Date'], 'Bullish', df.loc[i - 1, 'H'], df.loc[i + 1, 'L']))
        elif df.loc[i - 1, 'L'] > df.loc[i + 1, 'H']:  # Bearish OB
            ob_list.append((df.loc[i, 'Date'], 'Bearish', df.loc[i + 1, 'H'], df.loc[i - 1, 'L']))
    return ob_list

# Function to check trade confluences
def check_trade_conditions(df, ob_list):
    if df is None or ob_list is None or len(ob_list) == 0:
        return None
    for date, ob_type, high, low in ob_list:
        latest_price = df[df['Date'] == date]
        if latest_price.empty:
            continue
        latest_price = latest_price.iloc[0]
        price_above_vwap = latest_price['C'] > latest_price['VWAP']
        
        if ob_type == 'Bullish' and price_above_vwap:
            return 'BUY'
        elif ob_type == 'Bearish' and not price_above_vwap:
            return 'SELL'
    return None

# Function to execute trades instantly at market price
def place_trade(order_type, symbol, lot=0.1, sl_pips=50, tp_pips=100):
    tick_info = mt5.symbol_info_tick(symbol)
    if tick_info is None:
        print(f" Failed to get tick data for {symbol}. Symbol might be incorrect.")
        return

    # Market execution requires a price
    price = tick_info.ask if order_type == 'BUY' else tick_info.bid
    sl = price - sl_pips * 0.0001 if order_type == 'BUY' else price + sl_pips * 0.0001
    tp = price + tp_pips * 0.0001 if order_type == 'BUY' else price - tp_pips * 0.0001

    order = {
        "action": mt5.TRADE_ACTION_DEAL,
        "symbol": symbol,
        "volume": lot,
        "type": mt5.ORDER_TYPE_BUY if order_type == 'BUY' else mt5.ORDER_TYPE_SELL,
        "price": price,  #  FIXED: Added price for market execution
        "sl": sl,        #  FIXED: Added SL
        "tp": tp,        #  FIXED: Added TP
        "deviation": 10,
        "magic": 1001,
        "comment": "Institutional Order Flow Trade",
        "type_time": mt5.ORDER_TIME_GTC,
        "type_filling": mt5.ORDER_FILLING_IOC  #  FIXED: Changed to `ORDER_FILLING_IOC`
    }

    result = mt5.order_send(order)

    if result is None:
        print(f" Trade execution failed for {symbol}. Check logs.")
        print(f" Debug Order Data: {order}")
        return

    if result.retcode == mt5.TRADE_RETCODE_DONE:
        print(f" {order_type} order placed for {symbol} at {price}")
    else:
        print(f" Trade failed for {symbol}: {result.comment}")

# Function to track P/L
def track_performance():
    account_info = mt5.account_info()
    if account_info:
        print(f"Balance: {account_info.balance}, Equity: {account_info.equity}, P/L: {account_info.profit}")

# Main loop to run the bot for multiple symbols
symbols = ["EURUSD", "GBPUSD", "USDJPY", "XAUUSD", "USDCHF"]
timeframe = mt5.TIMEFRAME_M15

while True:
    for symbol in symbols:
        df = get_ohlc_data(symbol, timeframe)
        if df is not None:
            ob_list = detect_order_blocks(df)
            order_type = check_trade_conditions(df, ob_list)
            if order_type:
                place_trade(order_type, symbol)
            track_performance()
    time.sleep(60)  # Check every minute
```
